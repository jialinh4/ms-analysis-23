```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import wave
import scipy.io.wavfile
import numpy as np
```

```python
def find_filenames( path_to_dir, suffix=".csv" ):
    filenames = os.listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]
```

```python
path = 'C:\\Users\\Alkurdi\\Box\\2020_JUMP_ARCHES_anxiety\\data_analysis\\Hexoskin data\\'
folder = 'Rich\\record_232702\\'
# total time for rich 11/12/2021: 09:28:01 = 34081 seconds '

tf = 34081;
```

```python
def plot(path):
    fig, axs = plt.subplots(22, figsize = (12, 90))
    #fig = plt.figure(constrained_layout=True)
    fig.subplots_adjust(hspace=0.3)

    #wave file
    wav_files = find_filenames(path, '.wav')
    idx = 0
    for wav_file in wav_files:
        name = wav_file[:-3]
        wav_data = scipy.io.wavfile.read(path + wav_file)
        t_wav = np.linspace(0,tf,len(wav_data[1]))

        if len(wav_data[1])==0:
            print(name + ' is empty!')
        else:
            axs[idx].plot(t_wav, wav_data[1])
            axs.flat[idx].set(title=name)
            axs.flat[idx].set(xlabel='time (s)')
            
            idx+=1
    plt.savefig(path + '..//..//Hexoskin_VizFigures//' + str(folder).replace('\\', '_') + 'MultiModal_wav.png', dpi = 250)
    plt.show()
    plt.close()
    
    fig, axs = plt.subplots(9, figsize = (12, 35))
    #fig = plt.figure(constrained_layout=True)

    #csv file
    csv_files = find_filenames(path, '.csv')
    irrelevant_files = ['statistics.csv']
    csv_files = list(set(csv_files) - set(irrelevant_files))
    idx = 0
    for csv_file in csv_files:
        name = csv_file[:-3]
        csv_data = pd.read_csv(path + csv_file)
        
        if csv_data.empty:
            print(name + ' is empty!')
        else:
            axs[idx].plot(csv_data.iloc[:, 0], csv_data.iloc[:, 1])
            axs.flat[idx].set(title=name)
            axs.flat[idx].set(xlabel='time (s)')

            idx+=1
    plt.savefig(path + '..//..//Hexoskin_VizFigures//' + str(folder).replace('\\', '_') + 'MultiModal_csv.png', dpi = 250)
    plt.show()
```

```python
plot(path+ folder)
```
