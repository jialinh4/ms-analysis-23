```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import wave
import scipy.io.wavfile
```

```python
def find_filenames( path_to_dir, suffix=".csv" ):
    filenames = os.listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]
```

```python
path = 'C:\\Users\\Rachneet Kaur\\Box\\2020_JUMP_ARCHES_anxiety\\data_analysis\\Hexoskin data\\'
folder = 'Ayse_Hexoskin_102021\\record_231077-10-21-2021\\'
```

```python
def plot(path):
    fig, axs = plt.subplots(22, figsize = (15, 45))
    
    #wave file
    wav_files = find_filenames(path, '.wav')
    idx = 0
    for wav_file in wav_files:
        name = wav_file[:-3]
        wav_data = scipy.io.wavfile.read(path + wav_file)
        if len(wav_data[1])==0:
            print(name + ' is empty!')
        else:
            axs[idx].plot(wav_data[1])
            axs.flat[idx].set(ylabel=name)
            idx+=1
    plt.savefig(path + '..//..//Hexoskin_VizFigures//' + str(folder).replace('\\', '_') + 'MultiModal_wav.png', dpi = 250)
    plt.show()
    plt.close()
    
    fig, axs = plt.subplots(9, figsize = (15, 25))
    #csv file
    csv_files = find_filenames(path, '.csv')
    irrelevant_files = ['statistics.csv']
    csv_files = list(set(csv_files) - set(irrelevant_files))
    idx = 0
    for csv_file in csv_files:
        name = csv_file[:-3]
        csv_data = pd.read_csv(path + csv_file)
        if csv_data.empty:
            print(name + ' is empty!')
        else:
            axs[idx].plot(csv_data.iloc[:, 0], csv_data.iloc[:, 1])
            axs.flat[idx].set(ylabel=name)
            idx+=1
    plt.savefig(path + '..//..//Hexoskin_VizFigures//' + str(folder).replace('\\', '_') + 'MultiModal_csv.png', dpi = 250)
    plt.show()
```

```python
plot(path+ folder)
```
