```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

```python
path = 'C:\\Users\\Rachneet Kaur\\Box\\2020_JUMP_ARCHES_anxiety\\data_analysis\\E4 wristband data\\'
folder = 'A03A72_211112-184801'
csv_files = glob.glob(path + folder + '\\'+ "\*.csv")
```

```python
def plot(path):
    fig, axs = plt.subplots(8, figsize = (15, 15))
    
    #ACC file
    acc_file = pd.read_csv(path + 'ACC.csv')
    if acc_file.empty:
        print('ACC is empty!')
    else:
        axs[0].plot(acc_file.iloc[:, 0])
        axs.flat[0].set(ylabel='Acc x')
        axs[1].plot(acc_file.iloc[:, 1])
        axs.flat[1].set(ylabel='Acc y')
        axs[2].plot(acc_file.iloc[:, 2])
        axs.flat[2].set(ylabel='Acc z')
    
    #BVP file
    bvp_file = pd.read_csv(path + 'BVP.csv')
    if bvp_file.empty:
        print('BVP is empty!')
    else:
        axs[3].plot(bvp_file.iloc[:, 0])
        axs.flat[3].set(ylabel='BVP')
    
    #EDA file
    eda_file = pd.read_csv(path + 'EDA.csv')
    if eda_file.empty:
        print('EDA is empty!')
    else:
        axs[4].plot(eda_file.iloc[:, 0])
        axs.flat[4].set(ylabel='EDA')    

    #HR file
    hr_file = pd.read_csv(path + 'HR.csv')
    if hr_file.empty:
        print('HR is empty!')
    else:
        axs[5].plot(hr_file.iloc[:, 0])
        axs.flat[5].set(ylabel='HR') 

    #IBI file
    ibi_file = pd.read_csv(path + 'IBI.csv')
    if ibi_file.empty:
        print('IBI is empty!')
    else:
        axs[6].plot(ibi_file.iloc[:, 0], ibi_file.iloc[:, 1])
        axs.flat[6].set(ylabel='IBI') 

    #TEMP file
    temp_file = pd.read_csv(path + 'TEMP.csv')
    if temp_file.empty:
        print('TEMP is empty!')
    else:
        axs[7].plot(temp_file.iloc[:, 0])
        axs.flat[7].set(ylabel='Temp')
    
    plt.savefig(path + '..//E4_VizFigures//' + folder +'MultiModal.png', dpi = 250)
    plt.show()
```

```python
plot(path+ folder + '\\')
```
